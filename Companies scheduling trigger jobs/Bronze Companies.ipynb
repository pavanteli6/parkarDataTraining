```python
%sql
CREATE SCHEMA IF NOT EXISTS pavan_companies.bronze
     MANAGED LOCATION 'abfss://landing@parkaru15sa.dfs.core.windows.net/';  
```

```python
%sql
DROP TABLE IF EXISTS pavan_companies.bronze.companies;

CREATE TABLE pavan_companies.bronze.companies
AS
SELECT company_name, founded_date, country
  FROM read_files('abfss://landing@parkaru15sa.dfs.core.windows.net/companies/',
                  format => 'csv',
                  header => true);
                  
```
