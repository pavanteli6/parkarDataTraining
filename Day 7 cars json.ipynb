```python
df_car = spark.read.format("json").load("abfss://landing@parkaru15sa.dfs.core.windows.net/")
df_car_final = df_car.na.drop()
display(df_car_final)
```

```python
count_duplicates_before = df_car_final.count() - df_car_final.dropDuplicates().count()
df_car_final = df_car_final.dropDuplicates()
count_duplicates_after = df_car_final.count() - df_car_final.dropDuplicates().count()
display(spark.createDataFrame([(count_duplicates_before, count_duplicates_after)], ["duplicates_before", "duplicates_after"]))
display(df_car_final)
```
