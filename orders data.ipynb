```python
%sql
CREATE OR REPLACE TEMP VIEW tv_orders_fixed AS
SELECT
  value,
  regexp_replace(
    value,
    '"order_date":\\\\s*([0-9]{4}-[0-9]{2}-[0-9]{2})',
    '"order_date": "\\1"'
  ) AS fixed_value
FROM pavan_catalog_all.bronze.v_orders;
```
